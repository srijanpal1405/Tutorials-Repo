### **How to create basic arrays using Numpy?**

=> First we need to import the Numpy library to work with it.To import Numpy module we need to write -> import numpy.

But if we use "import numpy as np",  an alias for the namespace will be created. 

We can create Numpy ndarray object using array() function.

````python
#Creating a Numpy array

#Import the Library
import numpy as np

arr = np.array([1,2,3]) 

# This will print single dimensional array
print(arr)
````

We can pass list, tuple or array-like object in the array() method to create an ndarray.

````python
tup=(5,6,7,8)                  # Tuple
arr_tup=np.array(tup)          # Passing tuple in array function

print(arr_tup)                 # This will print 1-D array
````

We  can create array filled with zeros or ones.

````python
arr_0=np.zeros(4) # Using np.zeros() we can create an array of zeros. Here 4 is the number of elements (zeros)
arr_1=np.ones(7)  # Using np.ones() we can create an array of ones. Here 7 is the number of elements (ones)

print(arr_0)
print(arr_1)
````

We can create an array with a range of elements with the help of np.arange() function. We can also specify the first number, last number and step size.

````python
arr_1=np.arange(7)      # This will create an array starting from 0 to 6
arr_2=np.arange(2,9,2)  # This will create an array [2,4,6,8] as the starting point=2, ending point=9, step size=2

print(arr_1)
print(arr_2)
````

We can also use np.linspace() to create an array with values that are spaced linearly in a specified interval.

````python
arr1=np.linspace(0, 10, num=5)     # This will create an array between the range 0 to 10 and there is 5 equispaced elements in this range.
arr2=np.linspace(0, 10, num=6)     # This will create an array between the range 0 to 10 and there is 6 equispaced elements in this range.
print(arr1)
print(arr2)
````

### **Comparing Speed of Numpy Arrays and Lists**

Numpy arrays are 5-100 times faster than python lists.

````python
import time                               # Importing Time module

py_list=[i for i in range(10000000)]      # Python list --> [0,1,2,3,.....,9999999]

start=time.time()                         # starting time
py_list=[i+100 for i in py_list]          # iterating over python list through a for loop and adding 100 with each element
end=time.time()                           # ending time

elapsed_time=end-start                    # total time taken to complete the work (for loop iteration)

print(round(elapsed_time,6))              # printing the time elapsed upto 6 digit precision
````

````python
np_arr=np.array([i for i in range(10000000)])     # numpy array

start=time.time()
np_arr+=100                                       # adding 100 with each element in rthe array
end=time.time()

elapsed_time=end-start

print(round(elapsed_time,6))
````

From the above experiment you can see that numpy array traversal is 116.445 times (0.936337 / 0.008041) faster than python list. 

The reason for Numpy arrays being faster is -

* Numpy array is a collection of similar data-types that are densely packed in memory. A Python list can have different data-types, which puts lots of extra constraints while doing computation on it.
* Numpy is able to divide a task into multiple subtasks and process them parallelly.
* Numpy functions are implemented in C. Which again makes it faster compared to Python Lists.

### Illustration 1: NumPy Arrays are faster than Python lists
We will calculate the mean of a range of elements (50K to 1 million) in the array using both NumPy and lists. The arrays are randomly generated. A comparison plot between numpy arrays and lists will also be generated. 




````python
"""General comparison between NumPy and lists"""

import numpy as np
from time import time
import matplotlib.pyplot as plt


plot_x = []
numpy_plot = []
list_plot = []
number_of_elements = [x for x in range(50000, 1000000, 50000)]


for elements in number_of_elements:

  numpy_array = np.random.rand(elements)
  list_conv = list(numpy_array)

  #Start timing NumPy compuation
  start1 = time()

  #Compute the mean using NumPy
  numpy_mean = np.mean(numpy_array)

  #End timing
  end1 = time()

  #Time taken
  time1 = end1 - start1
  numpy_plot.append(time1)
  print(f"Computation time(NumPy): {time1}")

  #Start timing list computation
  start2 = time()

  #Compute the mean using lists
  list_mean = np.mean(list_conv)

  #End timing
  end2 = time()

  #Time taken
  time2 = end2 - start2
  list_plot.append(time2)
  print(f"Computation time(Python list): {time2}\n")
  print("."*50)


x = np.array(number_of_elements)
y = np.array(numpy_plot)
y1 = np.array(list_plot)
plt.plot(x, y, label="NumPy Array")
plt.plot(x, y1, label="Python lists")
plt.title("NumPy array v/s Python list")
plt.xlabel("Number of Elements in list/array")
plt.ylabel("Computation time in seconds(s)")
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.show()
````

# NumPy Tutorial : BROADCASTING

## Basic Definition and Terminology

In NumPy terminology, _broadcasting_ refers to how NumPy treats arrays with different shapes during arithmetic operations.

The smaller array is “broadcast” across the larger array so that they have compatible shapes (albeit certain constraints).

Now one might be wondering, **why broadcasting?**

Well, we've got you covered : 


````python
# Imports
import numpy as np
````

````python
a = np.array([1.0, 2.0, 3.0])
a
````

Suppose we wanted to update each element to twice its value.

Now we know, NumPy operations are usually done on pairs of arrays on an element-by-element basis. 
In the simplest case, the two arrays must have **exactly the same shape**. So we create another numpy array of the same size and dimensions as `a` and initialize it with all values as `2.0` . This is demonstrated in the snippet below : 

````python
b = np.array([2.0, 2.0, 2.0])
# Multiply a and b
a*b
````

While that clearly seems to work as expected, don't you wonder if we could find an alternate way that is a bit more concise, or in **python** terms, *more pythonic?*

Well, broadcasting is the way to go!
The simplest broadcasting example occurs when an array and a scalar value are combined in an operation :

````python
c = 2.0
# Multiply a and c
a*c
````

There you go. Everything in a single integer variable.

Now you might be wondering *what happened there?*
The result appears to be equivalent to the one using an array `b = [2.0 2.0 2.0]`. 
We can think of it as a scalar `c` being _stretched_ into an array of the same shape as `a` during the arithmetic operation. 
This has been beautifully visualized in the figure : 

NumPy Array stretching example

As [NumPy docs](https://numpy.org/doc/stable/user/basics.broadcasting.html) puts it : 
> The stretching analogy is only conceptual. NumPy is smart enough to use the original scalar value without actually making copies so that broadcasting operations are as memory and computationally efficient as possible.

Amazing, and _pythonic_ isn't it? And before we go into the thick of it, here's one more _pythonic_ example : 

````python
# slider that determines magnitude of scalar
s = 3 #@param {type:"slider", min:0.0, max:10.0, step:1.0}
````

Change the value of the slider to any integer value between 0 and 10, and following is your go-to array multiplier. Amazing, isn't it?

````python
a*s
````

And in case no one told you, multiplication wasn't the only arithmetic operation you could do with broadcasting. Let us conisder our primary array `a` and scalar `c`.

````python
# Addition
print(a+c,"\n")

# Subtraction
print(a-c,"\n")

# Subtraction with positions reversed
print(c-a)
````

Generally speaking, broadcasting provides an efficient means of vectorizing array operations so that looping occurs in C instead of python. But, as with all other useful things, broadcasting has its own constraints. And that is what we're going to discover in the following sections.

## Broadcastable Arrays

Now if broadcasting were the solution for all vectorization problems, why would we need the entire library of NumPy ;-;

So, let's figure out what 'audience' we can use this 'magic trick' on, or in NumPy terms, *which arrays are broadcastable.*  

In _NumPy_ terms, an array or a set of them is called **broadcastable** to a certain shape if the above rules provide a valid result.
_Also, you might like going back to the rules of broadcasting anytime you feel the need to._

That being said, let's take a few examples to have a better understanding : 

Suppose we have four NumPy arrays, `p`,`q`,`r` and `s` having the following shapes : 

`p.shape - (4,1)`

`q.shape - (1,3)`

`r.shape - (3,)`

`s.shape - ()` _(implying s is a scalar)_

Then, we can conclude `p`,`q`,`r` and `s` are all broadcastable to an array of shape (4,3) (i.e dimensions = 4x3), so that : 

- `p` acts like a (4,3) array where `p[:,0]` is broadcast to the other columns,

- `q` acts like a (4,3) array where `q[0,:]` is broadcast to the other rows,

- `r` acts like a (1,3) array and therefore like a (4,3) array where `r[:]` is broadcast to every row, and finally,

- `s` acts like a (4,3) array where the single value is repeated.

Let's have a glimpse at some examples that may help us understand better.

```
A      (2d array):  3 x 4
B      (1d array):      1
Result (2d array):  3 x 4

A      (2d array):  7 x 4
B      (1d array):      4
Result (2d array):  7 x 4

A      (3d array):  15 x 2 x 4
B      (3d array):  15 x 1 x 4
Result (3d array):  15 x 2 x 4

A      (3d array):  12 x 6 x 2
B      (2d array):       6 x 2
Result (3d array):  12 x 6 x 2

A      (3d array):  20 x 3 x 6
B      (2d array):       3 x 1
Result (3d array):  20 x 3 x 6
```

All of the broadcastable arrays above have aptly followed the rules for broadcasting. 

But just in case there are some avid fans of _the method of reductio ad absurdum_ (in simple words, proof by contradiction, let's see what happens when we try to broadcast arrays against the rules.

````python
# Array of shape (4,3)
against_the_rules_A = np.array([[1, 2, 3], [4, 7, 9], [3, 12, 21], [1, 0, 5]])

# Array of shape (2,1)
against_the_rules_B = np.array([[4], [7]])

try:
  against_the_rules_A*against_the_rules_B
except ValueError as ve:
  print(f'{ve}')
  pass
````

Just as we said, NumPy returns `ValueError` when trying to broadcast arrays that aren't broadcastable.

In the example above, **the shapes were neither equal nor consisted of only a single one of either shapes' dimensions as 1**. 
So, their fate was _inevitable_.

Here are some other examples of shapes that do not broadcast : 

```
A      (1d array):  4
B      (1d array):  5 # trailing dimensions do not match

A      (2d array):      2 x 1
B      (3d array):  7 x 3 x 4 # second from last dimensions mismatched
```

So, the "NOT-VALID" part being dealt with, let's see where we CAN use broadcasting.


1. Adding arrays of different dimension counts


````python
# 2D 
A_2D = np.array([[ 0.0,  5.0,  0.0],
           [15.0, 10.0, 25.0],
           [35.0, 20.0, 15.0],
           [30.0, 45.0, 30.0]])

# 1D
A_1D = np.array([3.0, 6.0, 9.0])

A_2D + A_1D
````

As shown in the figure alongside, A_1D is added to each row of A_2D. 

Stretching 1D array to 2D array

Quite fascinating, yet unsurprising by NumPy standards, isn't it?

Let's look at another one.

````python
p = np.array([5.0, 10.0, 25.0, 60.0])
q = np.array([3.0, 9.0, 6.0])

# Magic Code
p[:, np.newaxis] + q
````

Now you must be wondering, _what happened there?_

Well, say hello to the operator `newaxis`. 

It is an index operator which inserts a new axis into `p`, i.e it stretches its rows to produce as many columns as there are rows in `q`, thus producing identical sub-arrays.
